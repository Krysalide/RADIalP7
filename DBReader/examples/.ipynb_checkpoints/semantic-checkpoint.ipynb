```python
import json

def dump_pixano_2D_boxes(filename,boxes,task_name,path,imSize):
    anno = {}
    anno['task_name'] = task_name
    anno['annotations'] = []
    anno['data'] = {'type':'image','children':'','path':path}

    for i,bb in enumerate(boxes):
        x1 = bb[0]/imSize[1]
        y1 = bb[1]/imSize[0]
        x2 = bb[2]/imSize[1]
        y2 = bb[3]/imSize[0]
        
        box = {"id": str(i),
               "geometry": {
                    "vertices": [x1,y1,x2,y2],
                    "type": "rectangle"
                   },
                   "category": "car",
                   "options": {
                    "isMoving": "true"
                   }
                  }
        anno['annotations'].append(box)
        
    with open(filename, 'w') as outfile:
        json.dump(anno, outfile)
    outfile.close()
```

```python
def save_annotation_format(filename):

    template = {"name": "Object2D",
     "spec": {
      "plugin_name": "rectangle",
      "label_schema": {
       "category": [
        {
         "name": "car",
         "color": "red",
         "properties": [
          {
           "name": "isMoving",
           "type": "checkbox",
           "default": True
          }
         ]
        }
       ],
       "default": "car"
      },
      "data_type": "image"
     },
     "dataset": {
      "path": "Camera/",
      "data_type": "image"
     }
    }

    with open(filename, 'w') as outfile:
        json.dump(template, outfile)
    outfile.close()
```

```python
from scipy.spatial.transform import Rotation as R
def rotation3d(x,y,z,roll,yaw,pitch):
    
    pitch = np.radians(pitch)
    yaw = np.radians(yaw)
    roll = np.radians(roll)
    
    xyz = np.stack([x,y,z],axis=1)
    
    rotation_vector = np.array([roll,pitch,yaw])
    rotation = R.from_rotvec(rotation_vector)
    rotated_vec = rotation.apply(xyz)
    
    return rotated_vec[:,0],rotated_vec[:,1],rotated_vec[:,2]


#    float x,y,z;
#    float intensity;
#    float radialDistance;
#    float polAngle,aziAngle;
#    uint8_t layer_index;

def ApplyCalib(pcl,index,yaw):
    
    if(index%2):
        offset = -.50/180*3.14
    else:
        offset = +.10/180*3.14
    
    offset+=0.5/180*3.14
    
    x = pcl[:,4] * np.cos(pcl[:,5]+offset) * np.cos(pcl[:,6]+yaw)
    y = pcl[:,4] * np.cos(pcl[:,5]+offset) * np.sin(pcl[:,6]+yaw)
    z = pcl[:,4] * np.sin(pcl[:,5]+offset) + 0.37#0.42
    
    pcl[:,0] = x
    pcl[:,1] = y
    pcl[:,2] = z
    
    return pcl
```

```python
from sklearn.linear_model import RANSACRegressor
from sklearn.datasets import make_regression
from sklearn import linear_model

def remove_ground(xyz):
    
    id = np.where( (xyz[:,1]>-5) &  (xyz[:,1]<5) & (xyz[:,0]<15)& (xyz[:,0]>2))[0]
    X = xyz[id,0].reshape((-1, 1))
    y = xyz[id,2]
    lr = linear_model.RANSACRegressor()
    lr.fit(X,y)
    id = np.where( (xyz[:,2] > 0.1+(xyz[:,0] * lr.estimator_.coef_[0])))[0]
    
    return xyz[id,:]
```

```python
from sklearn.linear_model import RANSACRegressor
from sklearn.datasets import make_regression
from sklearn import linear_model

def remove_ground(xyz):
    
    id = np.where( (xyz1[:,1]>-5) &  (xyz1[:,1]<5) & (xyz1[:,0]<25)& (xyz1[:,0]>2))[0]
    
    
    X = xyz1[id,0].reshape((-1, 1))
    y = xyz1[id,2]
    lr = linear_model.RANSACRegressor()
    lr.fit(X,y)
    id = np.where( (xyz[:,2] > 0.1+(xyz[:,0] * lr.estimator_.coef_[0])))[0]
    
    return xyz[id,:]
```
