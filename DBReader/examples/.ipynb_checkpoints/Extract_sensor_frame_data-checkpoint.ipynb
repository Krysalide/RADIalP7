```python
from DBReader import SyncReader
from DBReader.SensorsReaders import CANDecoder
import matplotlib.pyplot as plt
```

### Open the dataset

```python
db = SyncReader('/media/julien/Storage_1/Dataset/HD_Radar/Calib_May28th/RECORD@2020-05-28_11.14.09',master='camera',tolerance=50000)
```

```python
# use print_info function to get all the details of the database content
db.print_info()
print('')

# Create an iterator on the dataset
ite = iter(db)
```

```python
# ALl data are return in a dictionay
data=next(ite)
print(data.keys())
```

```python
# and then each sensor returns a dictionary
print(data['camera'].keys())
print('')
```

### The camera is straight forward

```python
data=next(ite)
```

```python
print('the first camera timestamp: ',data['camera']['timestamp'],'us')
plt.imshow(data['camera']['data'])
```

### For the CAN, each sample is a group of 20 messages

```python
print(data['can'])
```

```python
# and to decode each message to get the real values, you need to call the can decoder and load a dbc file
can_decoder = CANDecoder('buick_enclave_2016.dbc')
messages = can_decoder.decode(data['can']['data'])
# messages contains a list of decoded can frames, each frame is composed of multiple signals
print(messages[0]['signals'].keys())
print(messages[0]['timestamp'])
```

### For the scala

```python
# There are 2 scala files, being scala1 and scala2
# This is the same sensor but it delivers 2 point clouds from the same scene, 
# one being more sentitive (lower threshold) than the other, but it has more
# noise
data=next(ite)
print(data['scala2'])
plt.figure(figsize=(8,16))
plt.plot(data['scala1']['data'][:,1],data['scala1']['data'][:,0],'r.')
plt.ylim(0,300)
plt.xlim(-20,20)
```
