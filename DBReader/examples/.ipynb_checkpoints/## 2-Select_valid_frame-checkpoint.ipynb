```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import cv2
import os
import glob
import pathlib
```

```python
root_dir = '/media/julien/HD_RADAR/FinalDataset/auto_anno'
files = os.listdir(root_dir)
anno={}
for f in files:
    anno[f]=-1

nbElt = len(files)

i=0
while i<nbElt:
    f = files[i]
    image= cv2.imread(os.path.join(root_dir,f))
   
    image = cv2.putText(image, '{}/{}'.format(i,nbElt), (50, 50), cv2.FONT_HERSHEY_SIMPLEX,1, (0, 0, 255), 2, cv2.LINE_AA)
    
    text = '????'
    if(anno[f]==0):
        text = 'delete'
    elif(anno[f]==1):
        text = 'valid'
        
    image = cv2.putText(image, text, (300, 50), cv2.FONT_HERSHEY_SIMPLEX,1, (0, 0, 255), 2, cv2.LINE_AA)


    cv2.imshow('annotation',image)
    k = cv2.waitKey(0) & 0xFF
    
    if k == 27:
        break
    elif k==83 or k==32: #forward
        i+=1
    elif k==81: #backard
        i-=1
    elif k== 118: #valid
        anno[f] = 1
        i+=1
    elif k== 100: #delete
        anno[f] = 0
        i+=1
        
    if(i<0):
        i=0
        
    if(not(i%50)):
        np.save('tmp_annotation.npy',anno)
        

cv2.destroyAllWindows()
```
