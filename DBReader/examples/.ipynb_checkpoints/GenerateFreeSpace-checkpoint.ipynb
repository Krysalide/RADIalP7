```python
def GetFreeSpace(pred,label):
    
    res = 103/512
    #imSize = (800,600) #103 par +/- 30m
    imSize = (512,512) #103 par +/- 30m
    
    # Select the class road
    mask = (pred==1).astype('uint8')
    
    # Extract the edges
    edges,_ = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    
    # Convert edges into polygon
    list_polygon = []
    for j in range(len(edges)):
        xy = []
        for i in range(len(edges[j])):
            #xy.append(edges[j][i][0])

            u = edges[j][i][0][0]*2
            v = edges[j][i][0][1]*2
            y,x = CAM.imageToWorld_Z(u,v,0)
            
            if(x<0):
                x = 999
            xy.append([y,x])

        if(len(xy)>=3):
            list_polygon.append(Polygon(xy))


    Multi = MultiPolygon(list_polygon)
    lineSplit = LineString([(-30,0), (-30,80),(30,80),(30,0),(-30,0)])
    freespace = split(Multi, lineSplit)

    # Smooth the freespace
    freespace2 = freespace.buffer(10, join_style=1).buffer(-10.0, join_style=1)
    
    # Translate by 30m lateral offset [+/-30m to 0-60m]
    freespace3 = translate(freespace2,103/2,0)
    
    # Upscale by a factor of 10, so that 1 pixel is 10cm
    freespace4 = scale(freespace3,512/103,512/103,0,origin=(0,0,0))
    
    # Build the pixel mask
    p = gpd.GeoSeries(freespace4)
    cartesian_map = rasterize(shapes=p,out_shape=imSize)
  
    # Then, transform to po;lar map
    polar_map,_=polarTransform.convertToPolarImage(cartesian_map,center=(cartesian_map.shape[1]/2,0),
                                             initialRadius=0,
                                            finalRadius=cartesian_map.shape[0],
                                            initialAngle=0,
                                            finalAngle=3.14,
                                            radiusSize=cartesian_map.shape[0],
                                            angleSize=180*5,#360*2,
                                            hasColor=False,
                                            order=0,
                                            border='constant',
                                            borderVal=0.0,
                                            useMultiThreading=False,
                                            settings=None,)

    polar_map = polar_map.transpose()
    
    # Finally, remove all the obstacles
    for i in range(label.shape[0]):
        x1 = label[i,1]
        y1 = label[i,2]
        x2 = label[i,3]
        y2 = label[i,4]

        dist = label[i,6]
        left,longi1 = CAM.imageToWorld_Z(x1,y2,0)
        right,longi2 = CAM.imageToWorld_Z(x2,y2,0)


        # convert in polar
        R = (np.sqrt(left*left+longi1*longi1) + np.sqrt(right*right+longi2*longi2))/2
        A1 = np.arctan(left/longi1)*180/np.pi
        A2 = np.arctan(right/longi2)*180/np.pi

        #A1 = np.clip(int(-A1*2 + 180),0,359)
        #A2 = np.clip(int(-A2*2 + 180),0,359)
        
        #A1 = np.clip(int(-A1*4 + 360),0,759)
        #A2 = np.clip(int(-A2*4 + 360),0,759)        
        
        #R = np.clip(int(R*10),0,799)
        
        A1 = np.clip(int(-A1*5 + 450),0,899)
        A2 = np.clip(int(-A2*5 + 450),0,899)
        R = np.clip(int(R*512/103),0,511)
        polar_map[R:, A2:A1]=0
        
    polar_map = np.fliplr(polar_map)
        
    return polar_map
```

```python
for index in range(num_samples):
    print(index)
    idx_in_table = dict_ids[index]
    label = labels[idx_in_table]
    # let s use the first label to read the data
    sample_id = label[0][0]
    
    image_filename = os.path.join(image_dir, "image_{:06d}.jpg".format(sample_id))
    img = cv2.imread(image_filename)

    #img = Image.open(image_filename).convert('RGB')
    
    with torch.no_grad():
        model = model.eval()
        img2 = transform(img).unsqueeze(0) # To tensor of NCHW
        img2 = img2.to(device)

        pred = model(img2).max(1)[1].cpu().numpy()[0] # HW

    id_road = np.where(pred==1)
    #id_obst = np.where((pred>=13) & (pred<=18))
    
    img[id_road[0],id_road[1],1] += 50
    #img[id_obst[0],id_obst[1],2] += 50
    
    img = np.clip(img,0,255)
    for box in label[:,1:5]:
        cv2.rectangle(img,tuple((int(box[0]/2),int(box[1]/2))),tuple((int(box[2]/2),int(box[3]/2))),(0,0,255),3)

        
    freespace_mask = GetFreeSpace(pred,label)
    
    target_filename = os.path.join(target_dir, "freespace_{:06d}.png".format(sample_id))
    cv2.imwrite(target_filename,freespace_mask*255)
    cv2.imshow('sample image',img)

    cv2.waitKey(100) 
    '''k = cv2.waitKey(0) # waits until a key is pressed
    if(k == ord('q')):
        break'''

cv2.destroyAllWindows()
```
