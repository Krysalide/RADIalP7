```python
import numpy as np
import cv2
import sys
from glob import glob
import os
```

```python
# Import DBReader
sys.path.append('/media/julien/Data1/pro/Dev/HD_Radar/HDRadar_dataset/DBReader')
from DBReader.DBReader import SyncReader,ASyncReader
from DBReader.SensorsReaders import CANDecoder
offset_radar = -180000
offset_scala = -40000

root_dir = '/media/julien/HD_RADAR1/FinalDataset/Pitch0Deg/'
target_folder = 'Step50m'

os.mkdir(target_folder)

Step = 50
```

```python
can_decoder = CANDecoder('buick_enclave_2016.dbc')
```

```python
def GenerateImages(dataset_name,output_folder,step):

    db = SyncReader(dataset_name,tolerance=20000,offset_radar=offset_radar,offset_scala=offset_scala)

    folder_name = dataset_name[dataset_name[:-1].rfind('/')+1:]
    
    oFolder = os.path.join(target_folder,folder_name)
    os.mkdir(oFolder)
    
    init = True
    absolute_dist = 0
    nbImage = 0
    timeRef=0
    

    for i in np.arange(len(db)-1,0,-1):

        datas=db.GetSensorData(i)
        img = datas['camera']['data']
        timestamp = datas['scala']['timestamp']
        SteeringWheel,YawRate,VehSpdAvgDrvn = db.GetMostRecentOdometry(can_decoder,timestamp)


        if(init):
            previousTime = timestamp
            stepDist = 0
            init=False
            absolute_dist=0
            timeRef=timestamp
            
            #cv2.imwrite(oFolder + 'image_{:06d}_dist_{:07.1f}m.png'.format(i,-absolute_dist),img)
            cv2.imwrite(oFolder + 'image_{:06d}_dist_{:07.1f}m_time_{:06d}.png'.format(i,-absolute_dist,0),img)
        else:
            dT_sec = (timestamp-previousTime)/1e6
            dist = VehSpdAvgDrvn*1000/3600 *dT_sec
            absolute_dist+=dist
            stepDist+=(-dist)
            previousTime = timestamp

            #if(stepDist>=step):
            if( (timestamp-timeRef)/1000>=2000)
                stepDist=0
                timeStep = int((timestamp-timeRef)/1000)
                timeRef = timestamp
                #cv2.imwrite(oFolder + 'image_{:06d}_dist_{:07.1f}m.png'.format(i,-absolute_dist),img)
                cv2.imwrite(oFolder + 'image_{:06d}_dist_{:07.1f}m_time_{:06d}.png'.format(i,-absolute_dist,timeStep),img)
```

```python
for folder in glob(os.path.join(root_dir,'*/')):
    # Check if it contains CAN data
    if(not len(glob(folder+'*can.bin'))):
        continue
        
    GenerateImages(folder,target_folder,20)
```
