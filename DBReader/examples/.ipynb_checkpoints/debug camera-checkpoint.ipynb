```python
from DBReader import ASyncReader
from DBReader.SensorsReaders import CANDecoder
import cv2
```

```python
db = ASyncReader('/media/julien/Storage_1/Dataset/HD_Radar/Dataset_1/RECORD@2020-06-02_13.50.58')
```

```python
print(db.getSensorFilters())
```

```python
db.setSensorFilters(['camera','can'])
```

```python
can_decoder = CANDecoder('buick_enclave_2016.dbc')
```

```python
can_decoder.db.messages
```

```python
# Create an iterator on the dataset
speed = 0
steering = 0
ite = iter(db)
stop = 0
for i in range(len(db)):
    timestamp,sample,sensor,data=next(ite)
    
    if(sensor=='camera'):
        font = cv2.FONT_HERSHEY_SIMPLEX
        image = cv2.resize(data,(640,480))
        cv2.putText(image, 'Speed {}'.format(speed), (20,20), font, 0.5, (0, 255, 0), 1, cv2.LINE_AA)
        cv2.putText(image, 'Steering {}'.format(steering), (20,40), font, 0.5, (0, 255, 0), 1, cv2.LINE_AA)
        cv2.imshow('frame',image)
        
    if(sensor=='can'):
        messages = can_decoder.decode(data)
        for m in messages:
            if 'WhlGrndVlctyLftDrvn' in m['signals']:
                speed = m['signals']['WhlGrndVlctyLftDrvn']
            if 'StrWhAng' in m['signals']:
                steering = m['signals']['StrWhAng']
     
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```
